```python
empresas_tickers = {
    "Repsol": "REP.MC",
    "Iberdrola": "IBE.MC",
    "Telefonica": "TEF.MC",
    "Santander": "SAN.MC",
    "BBVA": "BBVA.MC",
    "CaixaBank": "CABK.MC",
    "Bankinter": "BKT.MC",
    "Mapfre": "MAP.MC",
    "Acerinox": "ACX.MC"
}

tickers_empresas = {valor: clave for clave, valor in empresas_tickers.items()}
```

```python
data = {
    "Fecha": ["2020-01-01", "2020-12-31", "2021-11-10", "2022-12-31"],
    "Repsol": [1277396053, 1277396053, 1242396053, 1242396053],
    "Iberdrola": [1277396053, 1277396053, 1242396053, 1242396053],
    "Telefonica": [1277396053, 1277396053, 1242396053, 1242396053],
    "Santander": [1277396053, 1277396053, 1242396053, 1242396053],
    "BBVA": [1277396053, 1277396053, 1242396053, 1242396053],
    "CaixaBank": [1277396053, 1277396053, 1242396053, 1242396053],
    "Bankinter": [1277396053, 1277396053, 1242396053, 1242396053],
    "Mapfre": [1277396053, 1277396053, 1242396053, 1242396053],
    "Acerinox": [1277396053, 1277396053, 1242396053, 1242396053]
    }

# Crear DataFrame
num_acc = pd.DataFrame(data)
num_acc['Fecha'] = pd.to_datetime(num_acc['Fecha'])
num_acc.set_index('Fecha', inplace=True)

# Crear rango de fechas diario desde el inicio hasta el final
idx = pd.date_range(start=num_acc.index.min(), end=num_acc.index.max())

# Reindexar el DataFrame para incluir todas las fechas diarias, llenando los valores faltantes
num_acc = num_acc.reindex(idx, method='ffill')
```

```python
num_acc
```

```python
tickers = ["REP.MC", "IBE.MC", "TEF.MC", "SAN.MC", "BBVA.MC", "CABK.MC", "BKT.MC", "MAP.MC", "ACX.MC"]

# Descarga los datos de cierre ajustados
precios = yf.download(list(empresas_tickers.values()), start="2020-01-01", end="2022-12-31")['Adj Close']
```

```python
precios.columns = [tickers_empresas[ticker] for ticker in list(precios.columns)]
```

```python
# Asegurar el mismo orden de columnas en ambos DataFrames
orden_columnas = precios.columns
num_acc = num_acc[orden_columnas]
```

```python
num_acc = num_acc[num_acc.index.isin(precios.index)]
```
