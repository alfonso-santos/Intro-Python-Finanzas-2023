# Juego

Imagina que eres un inversor que quiere comprar acciones de una empresa.

* La empresa tiene un precio inicial de 100 euros por acción
* Cada día el precio puede subir o bajar un 10% según el mercado.
* Tu objetivo es comprar acciones cuando el precio sea bajo y venderlas cuando el precio sea alto, para maximizar tu beneficio.
* Tienes un presupuesto inicial de 1000 euros, y puedes comprar o vender hasta 10 acciones por día.
* El juego dura 10 días, y al final debes calcular tu beneficio o pérdida total.

Para implementar este juego en python, vamos a usar las siguientes estructuras de control:

- **if**: Para comprobar si el precio de la acción es mayor o menor que un cierto valor, y decidir si comprar o vender.
- **if-else**: Para ejecutar una acción u otra según una condición. Por ejemplo, si tenemos suficiente dinero para comprar, compramos; si no, no hacemos nada.
- **elif**: Para añadir más opciones a la estructura if-else. Por ejemplo, si el precio es muy bajo, compramos 10 acciones; si es bajo, compramos 5; si es normal, no hacemos nada; si es alto, vendemos 5; si es muy alto, vendemos 10.
- **for**: Para repetir una acción un número determinado de veces. Por ejemplo, para simular los 10 días del juego, usamos un bucle for que va desde 1 hasta 10.
- **while**: Para repetir una acción mientras se cumpla una condición. Por ejemplo, para pedir al usuario que introduzca un número válido de acciones a comprar o vender, usamos un bucle while que se repite hasta que el número sea entre 0 y 10.
- **break**: Para salir de un bucle antes de que termine. Por ejemplo, si el usuario introduce un número negativo de acciones, usamos break para terminar el bucle while y mostrar un mensaje de error.
- **continue**: Para saltar una iteración de un bucle y continuar con la siguiente. Por ejemplo, si el usuario introduce un número mayor que 10 de acciones, usamos continue para ignorar esa iteración y pedir otro número.

# Código completo

```python
from IPython.display import clear_output

# Importamos la librería random para generar números aleatorios
import random

# Definimos el precio inicial de la acción
precio = 100

# Definimos el presupuesto inicial del inversor
presupuesto = 1000

# Definimos el número inicial de acciones que posee el inversor
acciones = 0

# Definimos una función para calcular el beneficio o pérdida total
def calcular_beneficio(precio, presupuesto, acciones):
    return presupuesto + acciones * precio

# Empezamos el juego con un bucle for que va desde el día 1 hasta el día 10
for dia in range(1, 11):

    # Mostramos el día actual y el precio de la acción
    print(f"Día {dia}: El precio de la acción es {precio} euros.")

    # Refresh cell output
    clear_output(wait=True)

    # Comprobamos si el precio es menor o igual a cero
    if precio <= 0:
        # Si es así, mostramos un mensaje de que la empresa ha quebrado y salimos del bucle con break
        print("La empresa ha quebrado. Fin del juego.")
        break

    # Si no es así, continuamos con el juego
    else:
        # Usamos un bucle while para pedir y validar la entrada del usuario
        while True:
            # Pedimos al usuario que introduzca un número de acciones a comprar o vender
            numero = input("¿Cuántas acciones quieres comprar o vender?")

            # Usamos try-except para capturar posibles errores
            try:
                # Convertimos la entrada del usuario a un número entero
                numero = int(numero)
            except ValueError:
                # Si la entrada no es un número válido, mostramos un mensaje de error y usamos continue para pedir otro número
                print("Por favor, introduce un número válido.")
                continue

            # Comprobamos si el número es negativo
            if numero < 0:
                # Si es así, mostramos un mensaje de error y salimos del bucle con break
                print("No puedes introducir un número negativo.")
                break

            # Comprobamos si el número es mayor que 10
            elif numero > 10:
                # Si es así, mostramos un mensaje de advertencia y usamos continue para pedir otro número
                print("No puedes comprar o vender más de 10 acciones por día.")
                continue

            # Si el número es válido, salimos del bucle con break
            else:
                break

        # Comprobamos si el número de acciones a comprar o vender es positivo
        if numero > 0:
            # Usamos un bucle while para validar la entrada del usuario
            while True:
                # Pedimos al usuario que introduzca c para comprar o v para vender
                # Guardamos la entrada del usuario en una variable
                opcion = input("¿Quieres comprar o vender? Introduce c para comprar o v para vender.")

                # Comprobamos si la opción es c (comprar)
                if opcion == "c":
                    # Calculamos el coste de comprar el número de acciones indicado
                    coste = numero * precio

                    # Comprobamos si el presupuesto es suficiente para comprar
                    if presupuesto >= coste:
                        # Actualizamos el presupuesto y el número de acciones
                        presupuesto = presupuesto - coste
                        acciones = acciones + numero

                        # Mostramos un mensaje de confirmación
                        print(f"Has comprado {numero} acciones por {coste} euros. Tu presupuesto actual es {presupuesto} euros y tu número de acciones actual es {acciones}.")

                        # Salimos del bucle con break
                        break

                    # Si no es así, mostramos un mensaje de error y salimos del bucle con break
                    else:
                        print(f"No tienes suficiente dinero para comprar {numero} acciones. Tu presupuesto actual es {presupuesto} euros.")
                        break

                # Comprobamos si la opción es v (vender)
                elif opcion == "v":
                    # Calculamos el ingreso de vender el número de acciones indicado
                    ingreso = numero * precio

                    # Comprobamos si el número de acciones es suficiente para vender
                    if acciones >= numero:
                        # Si es así, actualizamos el presupuesto y el número de acciones
                        presupuesto = presupuesto + ingreso
                        acciones = acciones - numero

                        # Mostramos un mensaje de confirmación
                        print(f"Has vendido {numero} acciones por {ingreso} euros. Tu presupuesto actual es {presupuesto} euros y tu número de acciones actual es {acciones}.")

                        # Salimos del bucle con break
                        break

                    # Si no es así, mostramos un mensaje de error y salimos del bucle con break
                    else:
                        print(f"No tienes suficientes acciones para vender {numero}. Tu número de acciones actual es {acciones}.")
                        break

                # Si la opción no es ni c ni v, mostramos un mensaje de error y usamos continue para pedir otra opción
                else:
                    print("Por favor, introduce c o v.")
                    continue

        # Si el usuario no quiere comprar ni vender acciones, mostramos un mensaje de que no hace nada ese día
        elif numero == 0:
            print("No haces nada ese día.")

    # Generamos un número aleatorio entre -10 y 10 para simular la variación del precio
    variacion = random.randint(-10, 10)

    # Actualizamos el precio de la acción sumando la variación
    precio = precio + variacion

    # Mostramos una línea en blanco para separar los días
    print()

# Al final del juego, calculamos el beneficio o pérdida total usando la función definida anteriormente
beneficio = calcular_beneficio(precio, presupuesto, acciones)

# Mostramos el resultado final del juego
print(f"El juego ha terminado. El precio final de la acción es {precio} euros. Tu beneficio o pérdida total es {beneficio - 1000} euros.")
```
